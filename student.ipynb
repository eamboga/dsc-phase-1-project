To obtain data that would be significant to our analysis, we will drop all rows in TMDB that have less than 2000 votes or an evarage rating of less than 6.0.


```python
#Dropping rows where votes < 2000, or average vote_rating < 6.0
for row in TMDB_data.index:
    if TMDB_data.loc[row, 'vote_count'] < 2000 or TMDB_data.loc[row, 'vote_average'] < 6.0:
        TMDB_data.drop(row, inplace = True)
        
TMDB_data
```
